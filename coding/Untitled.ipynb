```python
def parse_date(date_str, value_date_str, bank_config):

    if "-" in date_str:
        date_format = bank_config["date_formats_24hr"]
        value_date_format = bank_config["value_date_format_dash"]

    elif "/" in date_str:
        # Convert the string to a datetime object
        date_format = bank_config["date_formats_12hr"]
        value_date_format = bank_config["value_date_format_slash"]

    date = datetime.strptime(date_str, '%m/%d/%Y %I:%M %p')
    value_date =  datetime.strptime(value_date_str, value_date_format)

    return date,value_date
```

```python
def convert_to_datetime(date_str, value_str):
    
    if '/' in date_str:
        try:
            # Attempt to parse as datetime with AM/PM format (MM/DD/YYYY hh:mm:ss AM/PM)
            date = datetime.strptime(date_str, "%d/%m/%Y %I:%M:%S %p")
            value_date = datetime.strptime(value_str, "%d/%m/%Y")
            return date, value_date
        except ValueError:
            try:
                # Attempt to parse as datetime with AM/PM format (MM/DD/YYYY)
                date = datetime.strptime(date_str, "%d/%m/%Y")
                value_date = datetime.strptime(value_str, "%d/%m/%Y")
                return date, value_date
            except ValueError:
                raise ValueError("Invalid input format")
                
    elif '-' in date_str:
        
        try:
            # Attempt to parse as datetime with time (DD-MM-YYYY hh:mm)
            date = datetime.strptime(date_str, "%d-%m-%Y %H:%M")
            value_date = datetime.strptime(value_str, "%d-%m-%Y")
            return date, value_date
            
        except ValueError:
            try:
                # Attempt to parse as datetime with time (DD-MM-YYYY)
                date = datetime.strptime(date_str, "%d-%m-%Y")
                value_date = datetime.strptime(value_str, "%d-%m-%Y")
                return date, value_date
            except ValueError:
                raise ValueError("Invalid input format")

```

```python
import pandas as pd

def clean_company_suffixes(company_name, suffixes):
    """Clean company name and apply suffix flags."""
    company_name = company_name.rstrip()
    
    for suf in suffixes:
        if company_name.lower().endswith(suf):
            cmp_name = company_name[:-len(suf)].strip()
            cleaned_name = cmp_name[0:-3] + cmp_name[-3:].replace("_","").replace(" ","").replace("-","").replace("()","")
            
            parking = True if suf in ["(parking)", "(parking )", "parking"] else False
            storage_space = True if (suf == "storage space" or suf == "(storage space)") else False
            rental = True if suf == "rental" else False
            seats = True if suf == "seats" else False
            
            return (cleaned_name, parking, rental, storage_space, seats)
    
    return (company_name, False, False, False, False)

def process_data(sheet_data, min_row, master_columns, clean_func):
    """Process sheet records for appropriate column names."""
    sheet_cols = [col.lower().strip() for col in sheet_data[min_row-1]]
    master_columns = [col.lower().strip() for col in master_columns]
    processed_data = []
    
    for data in sheet_data[min_row:]:
        processed_row = []
        
        for cr in range(len(master_columns)):
            if master_columns[cr] in sheet_cols:
                idx = sheet_cols.index(master_columns[cr])
                cleaned_data = clean_func(data[idx], ["(parking)", "(parking )", "parking", "storage space", "(storage space)", "rental", "seats"])
                processed_row.append(cleaned_data)
            else:
                processed_row.append((None, False, False, False, False))  # Handle if column not found
        
        processed_data.append(processed_row)
    
    return processed_data

# Example usage:
sheet_data = [
    ["Company Name", "Location", "Type"],
    ["ABC Inc (parking)", "New York", "Tech"],
    ["XYZ Ltd (storage space)", "San Francisco", "Retail"],
]

master_columns = ["company name","dummy", "parking", "rental", "storage space", "seats"]

processed_data = process_data(sheet_data, 1, master_columns, clean_company_suffixes)

# Print processed data
for row in processed_data:
    print(row)
```
